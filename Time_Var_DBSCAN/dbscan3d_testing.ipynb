```python
import cPickle as pickle, numpy as np, matplotlib.pyplot as plt
import psf# Module for gaussian PSF convolution

# Universal Simulation Parameters
sizeX,sizeY,sizeT = 10.,10.,48. # Define box size
dens = 10 #dens/sq deg
N_bg = dens*sizeX*sizeY # Number of background events to be distributed randomly
r_68 = 1.0 # Spatial PSF size in degrees 
numProcs = 12 # How many CPU to use


# Define a function to run the simulation given a set of input parameters
# Optimized via numpy's vector routines
def RunSim(N_sig_on, N_sig_off, t_burst):
    # Generate the signal
    X,Y,T = np.zeros(N_sig_on), np.zeros(N_sig_on), (np.random.ranf(N_sig_on)-.5)*t_burst
    X,Y,T = np.append(np.zeros(N_sig_off),X), np.append(np.zeros(N_sig_off),Y), np.append((np.random.ranf(N_sig_off)-.5)*sizeT,T)
    # Shift the signal photons using the PSF
    X,Y = psf.ApplyGaussianPSF(XMASTER=X,YMASTER=Y,r_68=r_68 )
    # Generate the background photons-Don't shift these by the PSF
    X = np.append(X,(np.random.ranf(N_bg)-.5)*sizeX)
    Y = np.append(Y,(np.random.ranf(N_bg)-.5)*sizeY)
    T = np.append(T,(np.random.ranf(N_bg)-.5)*sizeT)
    return X,Y,T
# Run a simulation with 500 photons distributed 
X,Y,T = RunSim(N_sig_on=200,N_sig_off=0, t_burst=6.)


from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(0,figsize=(12,12))
ax = fig.add_subplot(111, projection='3d',)

ax.scatter(X,Y,T, s=1)

plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('t')
import MCSTATS
expBG     = dens*np.pi*r_68**2 * (2./sizeT) # expected number of events in a cylinder of height 2a
nMin = expBG + 4.*np.sqrt(expBG)
scan = MCSTATS.DBSCAN_Compute_Clusters([(X,Y,T),], eps=r_68, timeScale=1, min_samples=nMin, numProcs=1,sigMethod='isotropic',inner=1.25, outer=2.0,plot=False,BGDensity=dens,TotalTime = sizeT)
        

for i in scan:
    if i.Coords!=0:
        for j in i.Coords:
            plt.xlim(-10,10)
            plt.ylim(-10,10)
            ax.set_zlim((-10,10))
            x,y,z = np.transpose(j)
            ax.scatter(x,y,z, s=3,color='r')
  

x=np.linspace(-.5,.5,500)
z=np.linspace(-1,1,2)
X, Z=np.meshgrid(x,z)
Y=np.sqrt(.25-X**2)

ax.plot_surface(X,Y,Z,color=(1,0,0,0.15),edgecolors=(1,0,0,.0))
ax.plot_surface(X,-Y,Z,color=(1,0,0,0.15),edgecolors=(1,0,0,.0))

ax.set_xlabel('Longitude [deg]',fontsize=14)
ax.set_ylabel('Latitude [deg]',fontsize=14)
ax.set_zlabel('T [mo]',fontsize=14)

ax.plot_surface(X-.5,Y-.75,Z-1,color=(0,0,1,0.15),edgecolors=(0,1,0,.0))
ax.plot_surface(X-.5,-Y-.75,Z-1,color=(0,0,1,0.15),edgecolors=(0,1,0,.0))

x=np.linspace(-1.5,1.5,500)
z=np.linspace(-3,3,2)
X, Z=np.meshgrid(x,z)
Y=np.sqrt(1.5**2-X**2)

ax.plot_surface(X,Y,Z,color=(0,1,0,.08),edgecolors=(0,1,0,.0))
ax.plot_surface(X,-Y,Z,color=(0,1,0,.08),edgecolors=(0,1,0,.0))

plt.xlim((-5,5))
plt.ylim((-5,5))
ax.set_zlim((-7,7))

plt.savefig('DBSCAN3D_demo.pdf')
```
