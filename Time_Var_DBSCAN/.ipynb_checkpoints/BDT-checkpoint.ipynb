```python
import cPickle as pickle, numpy as np, matplotlib.pyplot as plt
import psf# Module for gaussian PSF convolution

# Universal Simulation Parameters
sizeX,sizeY,sizeT = 30.,30.,24. # Define box size
dens = 10 #dens/sq deg
N_bg = dens*sizeX*sizeY # Number of background events to be distributed randomly
r_68 = 1.0 # Spatial PSF size in degrees 
numProcs = 12 # How many CPU to use
numSources = 1
numSims = 5000

def AddRandomPointSource(X,Y,T):
    # Pick random centroid
    cx,cy,ct = np.random.ranf(3)
    cx,cy,ct=(cx-0.5)*(sizeX-2),(cy-0.5)*(sizeY-2),(ct-0.5)*(sizeT-2)
    # pick burst length 
    t_burst = 6*np.random.ranf()
    if t_burst < 1: t_burst =1.
    # pick flux
    while True:    
        xs,ys = np.random.ranf(2)
        if xs<.05:
            N_sig_on = 20
            break
        elif 40**2*ys < xs**-2:
            N_sig_on = int(xs*400)
            break 
    #f.write(str(cx) + ' ' + str(cy) + ' ' + str(ct) + ' ' + str(N_sig_on) + ' ' + str(t_burst))
    x,y,t = np.zeros(N_sig_on), np.zeros(N_sig_on), (np.random.ranf(N_sig_on)-0.5)*t_burst
    x,y,t = x+cx, y+cy, t+ct
    # Shift the signal photons using the PSF
    x,y = psf.ApplyGaussianPSF(XMASTER=x,YMASTER=y,r_68=r_68 )
    X = np.append(X,x)
    Y = np.append(Y,y)
    T = np.append(T,t)
    
    return X,Y,T,(cx,cy,ct,N_sig_on,t_burst)

def AddCentralPointSource(X,Y,T):
    # Pick random centroid
    cx,cy,ct = 0,0,0
    # pick burst length 
    t_burst = 12*np.random.ranf()
    if t_burst < 1: t_burst =1.
    N_sig_on=0
    while True:    
        xs,ys = np.random.ranf(2)
        if xs<.1:
            N_sig_on = 10
        elif 10*ys < xs**-1:
            N_sig_on = int(xs*50)
        if N_sig_on*t_burst>20:break
        
    N_sig_on*=t_burst
    
    #f.write(str(cx) + ' ' + str(cy) + ' ' + str(ct) + ' ' + str(N_sig_on) + ' ' + str(t_burst))
    x,y,t = np.zeros(N_sig_on), np.zeros(N_sig_on), (np.random.ranf(N_sig_on)-0.5)*t_burst
    x,y,t = x+cx, y+cy, t+ct
    # Shift the signal photons using the PSF
    x,y = psf.ApplyGaussianPSF(XMASTER=x,YMASTER=y,r_68=r_68 )
    X = np.append(X,x)
    Y = np.append(Y,y)
    T = np.append(T,t)
    
    return X,Y,T,(cx,cy,ct,N_sig_on,t_burst)
    
    
# Define a function to run the simulation given a set of input parameters
# Optimized via numpy's vector routines

def RunSim(numSources):
    # Generate the signal
    #f = open("src.txt",'wb')
    src = []
    X,Y,T = [],[],[]
    for i in range(numSources):
        #ret = AddRandomPointSource(X,Y,T)
        ret = AddCentralPointSource(X,Y,T)
        X,Y,T,src2 = ret
        src.append(src2)
    #f.close()
    
    idx = np.where(np.logical_and(T>-.5*sizeT,T<0.5*sizeT))
    X= X[idx]
    Y= Y[idx]
    T= T[idx]
    
    # Generate the background photons-Don't shift these by the PSF
    X = np.append(X,(np.random.ranf(N_bg)-.5)*sizeX)
    Y = np.append(Y,(np.random.ranf(N_bg)-.5)*sizeY)
    T = np.append(T,(np.random.ranf(N_bg)-.5)*sizeT)
    
    return X,Y,T,src
# Run a simulation with 500 photons distributed 
mcSims, srcList = [], []
for i in range(numSims):
    X,Y,T,src = RunSim(numSources)
    mcSims.append((X,Y,T))
    srcList.append(src)
```

```python
def Identify_Clusters(scan,src):
    identities = []
    cx,cy,ct,N_sig_on,t_burst = np.transpose(src)
    for j in range(len(scan.Members)):
        # Compute radius in units of 3-sigma uncertainty on centroid positions
        r  = np.sqrt(np.square(scan.CentX[j]-cx) + np.square(scan.CentY[j]-cy))/(scan.MedR[j]/.5)
        dT = np.abs((scan.CentT[j]-ct)/(scan.Size95T[j]))
        idx = np.where(np.logical_and(r<1,dT<1)==True)[0]
        if np.shape(idx)[0] == 0:
            identities.append(-1)
        elif np.shape(idx)[0] == 1:
            identities.append(idx[0])
            #print "Cluster Identified With Original Source: ", idx
        else: 
            identities.append(idx[0])
            #print "Confusion Detected Defaulting to first source"
    return identities

# Identify detected clusters with the original list
identities = []
for i in range(len(scan)):
    identities.append(Identify_Clusters(scan[i],srcList[i]))

cnt = np.zeros(len(identities))
cnt10 = np.zeros(len(identities))

for k in range(len(cnt)):
    for i in range(numSources):
        if i not in identities[k]:
            if srcList[k][i][3]>20: cnt[k] +=1
            if srcList[k][i][3]>10: cnt10[k] +=1
        
           

print "Detected fraction of clusters:", np.mean([np.count_nonzero(np.array(k)!=-1) for k in identities])
print "Not detected with > 10 Photons:", np.mean(cnt10)
print "Not detected with > 20 Photons:", np.mean(cnt)
print "Mean Unidentified Clusters:",  np.mean([np.count_nonzero(np.array(k)==-1) for k in identities])


for i in range(len(scan[1].Members)):
    x,y,t = np.transpose(scan[1].Coords[i])
    plt.figure(0,figsize=(10,10))
    plt.scatter(x,y,s=.1)
    
```

```python
# Make list with half sims going to training and half going to testing
X,Y = [[],[],[],[],[],[],[]],  [[],[],[],[],[],[],[]]
for i in range(len(scan)):
    s = scan[i]
    sample = (s.Size95X,s.Size95Y,s.Size95T,s.MedR,s.MedT, s.Members, s.Sigs)
    for k in range(len(sample)):
        X[k] = np.append(X[k], sample[k]) # Nsamples x NFeatures
    Y = np.append(Y,(np.array(identities[i]) !=-1).astype(int)) # flag all identified clusters as signal and all others as background



cut = int(np.shape(X)[1]/2.)
X_train,Y_train = np.transpose(np.array(X)[::,:cut]),Y[:cut] # Training samples
X_test,Y_test = np.transpose(np.array(X)[::,cut:]),Y[cut:]   # BG samples

# Load BDT algorithms
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators=501, learning_rate=1.0,max_depth=3, random_state=0).fit(X_train, Y_train)

plt.figure(0,figsize=(10,14))
plt.subplot(421)
N = len(clf.feature_importances_)
ind = np.arange(N)+0.5    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
plt.bar(ind, clf.feature_importances_,   width, color='r')
plt.ylabel('Scores')
plt.title('BDT Feature Importance')
plt.xticks(ind+width/2., ('Size95X','Size95Y','Size95T','MedR','MedT','Members','Sigs') ,rotation=20)


print "Mean Classification Accuracy:", clf.score(X_test, Y_test)


pred = clf.predict(X_test)
real  = np.where(pred==1)[0]
fake = np.where(pred==0)[0]

def PlotHist(real,fake,f,l,subplot):
    plt.subplot(4,2,subplot)
    plt.xlabel(l)
    plt.hist(f[real],bins=np.linspace(np.min(f),np.max(f),25),color='b',histtype='step',normed=True,label='Real')
    plt.hist(f[fake],bins=np.linspace(np.min(f),np.max(f),25),color='r',histtype='step',normed=True,label='Fake')

X_test=np.transpose(X_test)    
PlotHist(real,fake,X_test[5],'Members',2)
plt.legend()
PlotHist(real,fake,X_test[0],'Size95X',3)
PlotHist(real,fake,X_test[1],'Size95Y',4)
PlotHist(real,fake,X_test[2],'Size95T',5)
PlotHist(real,fake,X_test[3],'MedR',6)
PlotHist(real,fake,X_test[4],'MedT',7)
PlotHist(real,fake,X_test[6],'Sigs',8)
    
#plt.figure(1,figsize=(10,6))
#ax1 = plt.subplot2grid((2,1), (0,0), colspan=2)
#plt.hist(np.transpose(clf.predict_proba(np.transpose(X_test)[real]))[0],bins=np.linspace(0,1,51),histtype='step',color='b',normed=True)
#plt.hist(np.transpose(clf.predict_proba(np.transpose(X_test)[fake]))[0],bins=np.linspace(0,1,51),histtype='step',color='r',normed=True)
#plt.yscale('log')
plt.savefig('bdt_hists.pdf')
```
