### The Monte Carlo

```python
from adsy.ipython import *
from pandas import *
extended_styles()
```

```python
import cPickle as pickle, numpy as np, matplotlib.pyplot as plt
import psf# Module for gaussian PSF convolution

# Universal Simulation Parameters
sizeX,sizeY,sizeT = 10.,10.,48. # Define box size
dens = 20 #dens/sq deg
N_bg = dens*sizeX*sizeY # Number of background events to be distributed randomly
r_68 = 1.0 # Spatial PSF size in degrees 
numProcs = 12 # How many CPU to use


# Define a function to run the simulation given a set of input parameters
# Optimized via numpy's vector routines
def RunSim(N_sig_on, N_sig_off, t_burst):
    # Generate the signal
    X,Y,T = np.zeros(N_sig_on), np.zeros(N_sig_on), (np.random.ranf(N_sig_on)-.5)*t_burst
    X,Y,T = np.append(np.zeros(N_sig_off),X), np.append(np.zeros(N_sig_off),Y), np.append((np.random.ranf(N_sig_off)-.5)*sizeT,T)
    # Shift the signal photons using the PSF
    X,Y = psf.ApplyGaussianPSF(XMASTER=X,YMASTER=Y,r_68=r_68 )
    # Generate the background photons-Don't shift these by the PSF
    X = np.append(X,(np.random.ranf(N_bg)-.5)*sizeX)
    Y = np.append(Y,(np.random.ranf(N_bg)-.5)*sizeY)
    T = np.append(T,(np.random.ranf(N_bg)-.5)*sizeT)
    return X,Y,T
# Run a simulation with 500 photons distributed 
X,Y,T = RunSim(N_sig_on=1000,N_sig_off=0, t_burst=1.)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(0,figsize=(10,4))
ax = fig.add_subplot(121, projection='3d')
ax.scatter(X,Y,T, s=2)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('t')


# Run a simulation with 5000 photons distributed 
X,Y,T = RunSim(N_sig_on=1000,N_sig_off=0,t_burst=100000)

ax = fig.add_subplot(122, projection='3d')
ax.scatter(X,Y,T, s=2)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('t')

plt.show()
```

```python
import DBSCAN,MCSTATS,dbscan_indexed
numSims = 6# number of simulations for each MC
#aList = np.logspace(0,1,20) # Scan search radius from 1-48 months. 
#N_sigList = range(0,101,25)

aList = np.logspace(0,1,8) # Scan search radius from 1-48 months. 
N_sigList = range(0,101,33)

burst_Time = 6. # temporal length of source in months

reload(MCSTATS)
reload(DBSCAN)

MS,MC = [],[]

for N_sig in N_sigList:
    mcSims=[RunSim(N_sig_on=N_sig,N_sig_off=0,t_burst=burst_Time) for k in range(numSims)]
    dbscanResults,sigs = [],[]
    
    for a in aList:
        bgDensity = N_bg/(sizeX*sizeY) # 2d mean background density
        expBG     = bgDensity*np.pi*r_68**2 * (2*a/sizeT) # expected number of events in a cylinder of height 2a
        #print "Search Radius=", a, " expected BG=", expBG , " TotalBG =", bgDensity*np.pi*r_68**2 
        nMin = expBG + 3*np.sqrt(expBG)
        if nMin<3.: nMin=3 # If the number of events is less than three, DBSCAN won't work very well....
         
        # Send results to DBSCAN.  timeScale tells DBSCAN to multiply the times by the argument.  
        scan = MCSTATS.DBSCAN_Compute_Clusters(mcSims, eps=r_68,min_samples=nMin, timeScale=float(r_68)/a,numProcs=numProcs,plot=False)
        
        dbscanResults.append(scan)
        # Use annulus method here.
        #sigs.append(MCSTATS.Cluster_Sigs_3d_Annulus(scan,mcSims,inner=1.25,outer=2.0))
        sigs.append(MCSTATS.Cluster_Sigs_BG_3d(scan,BGDensity=N_bg/float(sizeX*sizeY), totalTime = sizeT))
    # Compute averages 
    counts,sigmas = [],[]
    for b in sigs:
        count,avgSig = 0., 0.
        for s in b:
            count+=len(list( np.where((np.array(s)>2.)==True)[0] ))
            if len(s)>0:
                avgSig += np.max(s)
        counts.append(count/float(len(b)))
        sigmas.append(avgSig/float(len(b)))
        
    MS.append(sigmas)
    MC.append(counts)
#print "N_min=", nMin 
```

```python
plt.figure(0,figsize=(10,4))
maxSig = []

for i in range(len(N_sigList)):
    N_sig,sigmas,counts = N_sigList[i],MS[i],MC[i] 
    plt.subplot(121)
    if N_sig==0:plt.plot(aList,sigmas,ls='--')
    else:plt.plot(aList,sigmas)
    #plt.vlines(bgDensity * 4./3.* np.pi*r_68**3 , .1,100,label='Expected BG Count=1',linestyle='--') # expected number of events in a cylinder that circumscribes  )
    plt.xlabel(r'Temporal Search Radius a [mo]')
    plt.ylabel(r'Average Max Significance $\sigma$')
    #plt.yscale('log')
    #plt.xscale('log')
    
    plt.subplot(122)
    if N_sig==0: plt.plot(aList,counts,label=r'BG Only',ls='--')
    else:plt.plot(aList,counts,label=r'$N_{sig}^{on}=$' + str(N_sig) + r"$\gamma$")
        
    plt.ylabel(r'Detected Clusters $N_{clusters}$')
    plt.xlabel(r'Temporal Search Radius a [mo]')
    #plt.yscale('log')
    #plt.xscale('log')
    
    if aList[np.argmax(sigmas)]>1: maxSig.append(aList[np.argmax(sigmas)])

plt.legend(ncol=2,bbox_to_anchor=(2.3, 1.00))
        
plt.show()    
```

Above, we plot the average maximum cluster significance (left) and number of clusters (right) as a function of the DBSCAN scale factor \\(a\\) for several different values of \\(N_{sig}^{on}\\).  We see that the region that maximizes significance is coincident with the region that detect only one cluster (no false clusters detected).  The region also appears to be stable over a reasonably large range of scale factors.  

Furthermore, the maximum significance occurrs at the theoretical maximum (within the granularity of the simulations).  This maximium is independent of the number of signal photons and is found by choosing the scale factor which forms the most spherical cluster (may want to correct metric to be cylindrical rather than ellipsoidal).  For a PSF of 1 deg. radius, this gives

$$a = \frac{1}{t_{burst}} \frac{\Delta T}{\Delta X}=1~month \times \frac{48}{10}=4.8~month/deg.$$

Fixing \\( N_{\sigma}=3, N^{sig}_{on} = 100 , N^{BG} = 1000 \\), we now look at scale factor vs signal burst length to see if there is a correlation.  To some extent, we expect the distribution to plateau over a rage of scale factors and begin declining once the temporal search radius is larger than the burst length/2.

```python
numSims = 6 # number of simulations for each MC
aList = np.linspace(1,8,10) # scale factors to scan over
bursts = np.linspace(0,12,10)

MS,MC = np.zeros(shape=(len(aList), len(bursts))) ,np.zeros(shape=(len(aList), len(bursts)) )

for i in range(len(aList)):
    for j in range(len(bursts)):
        
        mcSims=[RunSim(N_sig_on=100*bursts[j]/6.,t_burst=bursts[j], N_sig_off =0 ) for l in range(numSims)]
        dbscanResults,sigs = [],[]
        bgDensity = N_bg/float(sizeX*sizeY) # 2d mean background density
        expBG     = bgDensity*np.pi*r_68**2 * (2.*aList[i]/sizeT) # expected number of events in a cylinder of height 2a
        #print "Search Radius=", a, " expected BG=", expBG , " TotalBG =", bgDensity*np.pi*r_68**2 
        nMin = expBG + 3.*np.sqrt(expBG)
        if nMin<3.: 
            nMin=3# If the number of events is less than three, DBSCAN won't work very well....
            print nMin
            
        scan = MCSTATS.DBSCAN_Compute_Clusters(mcSims, eps=r_68,min_samples=nMin, timeScale=float(r_68)/aList[i],numProcs=numProcs,plot=False)
        dbscanResults.append(scan)
        #sigs.append(MCSTATS.Cluster_Sigs_3d_Annulus(scan,mcSims,inner=1.25,outer=2.0))
        sigs.append(MCSTATS.Cluster_Sigs_BG_3d(scan,BGDensity=N_bg/float(sizeX*sizeY), totalTime = sizeT))
        # Compute averages 
        counts,sigmas = [],[]
        for a in sigs:
            count,avgSig = 0., 0.
            for s in a:
                count+=len(s)
                if len(s)>0:
                    avgSig += np.max(s)
            MC[i][j] = count/float(len(a))
            MS[i][j] = avgSig/float(len(a))
```

```python
plt.figure(0,figsize=(12,4))
plt.subplot(121)
cs = plt.contourf(bursts,aList,MS,levels=np.linspace(0,8,61))
plt.xlabel(r'$T_{burst}$')
plt.ylabel(r'Temporal Search Radius $a$')
cbar = plt.colorbar(cs)
plt.title('Average Max Significance $\sigma$')

plt.subplot(122)
cs = plt.contourf(bursts,aList,MC,levels=np.linspace(0,10,61))
cbar = plt.colorbar(cs)
cs = plt.contour(bursts,aList,MC,levels=[0,.75,1,2],colors='k')
plt.clabel(cs)

plt.xlabel(r'$T_{burst}$')
plt.title('Average Number of Clusters $N_{clusters}$')

plt.show()
```

Remarkably, there is no dependence of the scale factor on the burst time, yet the scale factor is clearly quite sensitive to the what is likely the signal to noise ratio.  It seems that once the scale factor is chosen w.r.t. the SNR, the points are able to link together anyway even if the cluster is actually larger.  This is an advantage because we can resolve any length burst with a single, minimal choice of scale factor, provided that burst surpasses a minimal signal to noise ratio.
