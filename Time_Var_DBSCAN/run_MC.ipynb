## Tuning DBSCAN for Transient and Time Variabile Sources

### The Monte Carlo

```python
from adsy.ipython import *
from pandas import *
extended_styles()
```

The goal of this notebook is to test DBSCAN's ability to resolve time variable sources in an isotropic and homogenous background, and to determine the algorithms detectability threshold and statistical significance compared to simple binning.  With an eye on time variability at the galactic center, this will also serve as a case study to detect low significance, time variable point sources that are otherwise lost in the full time-integrated signal.  This may offer a method to expand, for example, 2FGL into the time domain.  

To demonstrate viability we will write a Monte-Carlo which produces a single isolated source in the center of the simulation volume.  The signal template will simply be cylindrically symmetric, with the axis pointing temporally, and the radial (spatial) events distributed according to a Gaussian with 68% containment radius matching the average of the FERMI-LAT performance specifications for a given energy range.  (Found at:http://www.slac.stanford.edu/exp/glast/groups/canda/lat_Performance.htm)  For the test case we will assume this is 1 degree. 

As input parameters we will have the number of signal photons when the burst is on/off, \\( N_{on}^{sig}, N_{off}^{sig} \\), the sources burst-time \\(t_{burst}\\), the scale-factor for the time component of DBSCAN's metric \\(a\\).

Let us first generate the simulation data.  Later we will discuss the DBSCAN aspects.  In this code block we generate a 10x10 (deg) x48 (month) box.  We then have a background count specified by \\(N_{bg}=1000\\).  For now, we also assume \\( N_{off}^{sig}=0 \\) so that the source is isolated in time.  The source is then placed at the center of the box, with \\( N_{on}^{sig}\\) photons distributed isotropically over a time \\(t_{burst}\\).

```python
import cPickle as pickle, numpy as np, matplotlib.pyplot as plt
import psf# Module for gaussian PSF convolution

# Universal Simulation Parameters
sizeX,sizeY,sizeT = 10.,10.,48. # Define box size
dens = 25 #dens/sq deg
N_bg = dens*sizeX*sizeY # Number of background events to be distributed randomly
r_68 = 1.0 # Spatial PSF size in degrees 
numProcs = 12 # How many CPU to use


# Define a function to run the simulation given a set of input parameters
# Optimized via numpy's vector routines
def RunSim(N_sig_on, N_sig_off, t_burst):
    # Generate the signal
    X,Y,T = np.zeros(N_sig_on), np.zeros(N_sig_on), (np.random.ranf(N_sig_on)-.5)*t_burst
    X,Y,T = np.append(np.zeros(N_sig_off),X), np.append(np.zeros(N_sig_off),Y), np.append((np.random.ranf(N_sig_off)-.5)*sizeT,T)
    # Shift the signal photons using the PSF
    X,Y = psf.ApplyGaussianPSF(XMASTER=X,YMASTER=Y,r_68=r_68 )
    # Generate the background photons-Don't shift these by the PSF
    X = np.append(X,(np.random.ranf(N_bg)-.5)*sizeX)
    Y = np.append(Y,(np.random.ranf(N_bg)-.5)*sizeY)
    T = np.append(T,(np.random.ranf(N_bg)-.5)*sizeT)
    return X,Y,T
# Run a simulation with 500 photons distributed 
X,Y,T = RunSim(N_sig_on=1000,N_sig_off=0, t_burst=1.)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(0,figsize=(10,4))
ax = fig.add_subplot(121, projection='3d')
ax.scatter(X,Y,T, s=2)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('t')


# Run a simulation with 5000 photons distributed 
X,Y,T = RunSim(N_sig_on=1000,N_sig_off=0,t_burst=100000)

ax = fig.add_subplot(122, projection='3d')
ax.scatter(X,Y,T, s=2)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('t')

plt.show()
```

As a reality check to ensure our MC is correctly simulating, we plot a few example cases above to ensure we have a centered, cylindrically symmetric source with an axis in the time direction.  These look correct so we will now attempt to profile DBSCAN in 3-dimensions.

### Tuning DBSCAN for 3-Dimensions 

We wish to determine three things:

1. What scale factor for the time coordinate is relevant for a cluster with a known length? In other words, once we know the time variability of our source, we need to choose the appropriate metric for DBSCAN which maximizes the statistical significance of the cluster.  In a general search algorithm, we will likely need to scan over a range of scale factors and a transient source would then show up as a resonance in a plot of scale factor \\(a\\) vs cluster significance \\(\sigma\\).

2. How stable is the algorithm under changes in DBSCAN parameters for a real source? We would like to understand whether or not our algorithm is sensitive to small changes in \\(N_{min}\\) or \\(a\\).  The less sensitivity exhibited, the more robust the results will be.  In addition, it will allow for less granularity when scanning over the scale factor, resulting in a much more (computationally) efficient search algorithm.

3. What is the detection threshold in terms of overdensity compared to the backgrounds and how do the significances change when we lower \\(N_{min}\\) or change \\(a\\)?  From previous 2-dimensional studies of these properties, it is generally better to begin with \\(N_{min}=N_{bg}^{exp.}+3 \sqrt{N_{bg}^{exp.}}\\) as the threshold, where \\(N_{bg}^{exp.}\\) is the expected mean background count for a given epsilon, but this decreases the number of detectable real clusters.  We will characterize this again for the 3-d case.

Let us begin to answer these quesetions by simulating a source with \\(t_{burst}=1\\) month over the 4-year Fermi-LAT mission lifetime.

```python
plt.figure(0,figsize=(10,14))

def plot_q(aList, q, subplot, ylabel,label):
    plt.subplot(3,2,subplot)
    if N_sig==0:plt.plot(aList,q,ls='--',label=label)
    else:plt.plot(aList,q,label=label)
    plt.xlabel(r'Temporal Search Radius a [mo]')
    plt.ylabel(ylabel)
    plt.xscale('log')
    #plt.yscale('log')

def plot_all(N_sigList,label):
    for i in range(len(N_sigList)):
        matplotlib.rcParams.update({'font.size': 12})
        N_sig = N_sigList[i]
        plot_q(aList,MS[i],1,r'Average Max Significance $\sigma$',label+ str(np.round(N_sig,2)))
        plt.legend(ncol=2,bbox_to_anchor=(2.2, .85))
        plot_q(aList,MC[i],5,r'Detected Clusters $N_{clusters}$ s>2',label)
        plot_q(aList,2.*PD[i],3,r'Cluster Radius R',label)
        plot_q(aList,4.*PT[i],4,r'Cluster Length T',label)
        plot_q(aList,MM[i],6,r'Cluster Members',label)


plot_all(N_sigList,label = r'$N_{\mathrm{sig-on}}=$')
plt.savefig('stats_vary_signal.pdf')
```

We see above that there is a tradeoff.  While detection efficiency for weak clusters is increased at higher cluster radii, the clusters gain slightly higher significances at lower a.  However, at a=1, the background contribution is highly suppressed (for N_sigma=5).  We see stability in the counts, temporal size, significance of the properties over a=1-3, while they converge for reasonable numbers of signal photons near a=1.

It appears that there is little dependence on the length of bursts in fixing the search radius.  We will now check the relationship between search radius and background level.

```python
numSims = 6# number of simulations for each MC
aList = np.logspace(0,1,8) # Scan search radius from 1-48 months. 
SNR = np.linspace(.1,1,5)
MS,MC = [],[]

for snr in SNR:
    mcSims=[RunSim(N_sig_on=0.,N_sig_off=0,t_burst=6) for i in range(numSims)]
    dbscanResults,sigs = [],[]
    
    for a in aList:
        N_bg = dens*sizeX*sizeY
        bgDensity = N_bg*snr/float((sizeX*sizeY)) # 2d mean background density
        expBG     = bgDensity*np.pi*r_68**2 * (2*a/sizeT) # expected number of events in a cylinder of height 2a
        #print "Search Radius=", a, " expected BG=", expBG , " TotalBG =", bgDensity*np.pi*r_68**2 
        nMin = expBG + 3*np.sqrt(expBG)
        if nMin<3.: nMin=3 # If the number of events is less than three, DBSCAN won't work very well....
         
        # Send results to DBSCAN.  timeScale tells DBSCAN to multiply the times by the argument.  
        scan = MCSTATS.DBSCAN_Compute_Clusters(mcSims, eps=r_68,min_samples=nMin, timeScale=float(r_68)/a,numProcs=numProcs,plot=False)
        
        dbscanResults.append(scan)
        # Use annulus method here.
        #sigs.append(MCSTATS.Cluster_Sigs_3d_Annulus(scan,mcSims,inner=1.25,outer=2.0))
        sigs.append(MCSTATS.Cluster_Sigs_BG_3d(scan,BGDensity=N_bg/float(sizeX*sizeY), totalTime = sizeT))
    # Compute averages 
    counts,sigmas = [],[]
    for b in sigs:
        count,avgSig = 0., 0.
        for s in b:
            count+=len(list( np.where((np.array(s)>2.)==True)[0] ))
            if len(s)>0:
                avgSig += np.max(s)
        counts.append(count/float(len(b)))
        sigmas.append(avgSig/float(len(b)))
        
    MS.append(sigmas)
    MC.append(counts)
#print "N_min=", nMin 
```

```python
plt.figure(0,figsize=(10,4))
maxSig = []

for i in range(len(SNR)):
    snr,sigmas,counts = SNR[i],MS[i],MC[i] 
    plt.subplot(121)
    if snr==0: plt.plot(aList,counts,label=r'BG Only',ls='--')
    else:plt.plot(aList,sigmas)
    #plt.vlines(bgDensity * 4./3.* np.pi*r_68**3 , .1,100,label='Expected BG Count=1',linestyle='--') # expected number of events in a cylinder that circumscribes  )
    plt.xlabel(r'Temporal Search Radius a [mo]')
    plt.ylabel(r'Average Max Significance $\sigma$')
    #plt.yscale('log')
    #plt.xscale('log')
    
    plt.subplot(122)
    if snr==0: plt.plot(aList,counts,label=r'BG Only',ls='--')
    else:plt.plot(aList,counts,label=r'BGLEVEL=' + str(snr))
        
    plt.ylabel(r'Detected Clusters $N_{clusters}$')
    plt.xlabel(r'Temporal Search Radius a [mo]')
    #plt.yscale('log')
    #plt.xscale('log')
    
    if aList[np.argmax(sigmas)]>1: maxSig.append(aList[np.argmax(sigmas)])

plt.legend(ncol=2,bbox_to_anchor=(2.3, 1.00))
        
plt.show()    
```

Remarkably, there is no dependence of the scale factor on the burst time, yet the scale factor is clearly quite sensitive to the what is likely the signal to noise ratio.  It seems that once the scale factor is chosen w.r.t. the SNR, the points are able to link together anyway even if the cluster is actually larger.  This is an advantage because we can resolve any length burst with a single, minimal choice of scale factor, provided that burst surpasses a minimal signal to noise ratio.
